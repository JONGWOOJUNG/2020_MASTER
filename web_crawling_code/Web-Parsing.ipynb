```python
df1 = pd.read_csv('서울특별시 공공자전거 대여이력 정보_2018년_1분기1.csv',encoding='cp949')
print(df1.shape)
df1.head()
```

```python
def func1(data):
    result = data[2:-1]
    return result

df1['start'] = df1["'대여대여소명'"].apply(func1)
df1['end'] = df1["'반납대여소명'"].apply(func1)
df1[['start','end']].head()
```

```python
list1 = df1['start'].unique().tolist()
```

```python
# 이동 정보 
minu_list = []
kilo_list =[]
kalori_list = []
minu_MIN_list =[]
kilo_MIN_list =[]
kalori_MIN_list = []
#출발 도착지 정보 
start_list = []
end_list =[]
```

```python
def collect_data(driver,list1):
    driver.find_element_by_xpath('//*[@id="search.tab2"]/a').click()
    for i in list1: 
        try:        
            # 출발지 입력 
            time.sleep(1)
            driver.find_element_by_xpath('//*[@id="info.route.waypointSuggest.input0"]').send_keys('서울역')
            time.sleep(1)
            driver.find_element_by_xpath('//*[@id="info.route.waypointSuggest.input0Suggest"]/div/div[1]/div/ul/li[1]/a').click()

            # 도착지 입력
            time.sleep(1)
            driver.find_element_by_xpath('//*[@id="info.route.waypointSuggest.input1"]').send_keys(i)
            time.sleep(1)
            driver.find_element_by_xpath('//*[@id="info.route.waypointSuggest.input1Suggest"]/div/div[1]/div/ul/li[1]/a').click()

            time.sleep(2)

            # 데이터 수집 
            driver.find_element_by_xpath('//*[@id="biketab"]').click()

            time.sleep(1)
            minu = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[1]/div[1]/div[1]/p/span[1]').text
            minu_MIN = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[2]/div[1]/div[1]/p/span[1]').text

            kilo = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[1]/div[1]/div[1]/p/span[2]/span[1]').text
            kilo_MIN = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[2]/div[1]/div[1]/p/span[2]/span[1]').text

            kalori = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[1]/div[1]/div[1]/div/span[2]').text
            kalori_MIN = driver.find_element_by_xpath('//*[@id="info.bikeRoute"]/div[1]/ul/li[2]/div[1]/div[1]/div/span[2]').text

            # 데이터 리스트 추가 
            minu_list.append(minu)
            kilo_list.append(kilo)
            kalori_list.append(kalori)
            minu_MIN_list.append(minu_MIN)
            kilo_MIN_list.append(kilo_MIN)
            kalori_MIN_list.append(kalori_MIN)
            start_list.append('서울역')
            end_list.append(i)

            print("Done")

            driver.find_element_by_xpath('//*[@id="search.tab2"]/a').click()

        except Exception as e :
            print("ERROR 발생 ")
            print(e)
            driver.find_element_by_xpath('//*[@id="search.tab2"]/a').click()

        finally:
            pass
    return driver, list1
```

```python
collect_data(driver,list1)
```

```python
data = pd.DataFrame({"출발지":start_list,"도착지":end_list, "자전거도로우선_걸린시간":minu_list, 
              "자전거도로_거리":kilo_list, "자전거도로_소모량":kalori_list, 
              '최단시간':minu_MIN_list, "최단거리":kilo_MIN_list,"최단소모량":kalori_MIN_list}).tail(15)
```

```python
def func2(row):
    if len(row)>4:
        result = row.split('시간')
        result = int(result[0])*60 + int(result[1][:-1])
        return result 
    else:
        return (row[:-1])
    
data['자전거도로우선_걸린시간_clean'] = data['자전거도로우선_걸린시간'].apply(func2)
data['최단시간_clean'] = data['최단시간'].apply(func2)
```
